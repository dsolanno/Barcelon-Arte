```python
#Removing Nans from data
arquitectura_cleaned = arquitectura.dropna(inplace=False)
cartells_cleaned = cartells.dropna(inplace=False)
dibuix_cleaned = dibuix.dropna(inplace=False)
escultura_cleaned = escultura.dropna(inplace=False)
estampes_cleaned = estampes.dropna(inplace=False)
medalles_cleaned = medalles.dropna(inplace=False)
moda_cleaned = moda.dropna(inplace=False)
pintura_cleaned = pintura.dropna(inplace=False)
vitrall_cleaned = vitrall.dropna(inplace=False)
```

```python
vitrallSubset = vitrall.dropna(inplace=False, subset=['Autor', 'Inici', 'Coordenades'])
```

```python
vitrallOutput=vitrallSubset[['Autor', 'Inici', 'Coordenades']]
coordinates = vitrallOutput['Coordenades'].str.split(' ')
coordsSplitted = pd.DataFrame({'Lat': coordinates.str[0], 'Lon': coordinates.str[1]})
coordsSplitted
```

```python
vitrallOutput['Latitude'] = coordsSplitted.Lat
vitrallOutput['Longitude'] = coordsSplitted.Lon
vitrallOutput['Coordenades'] = vitrallOutput['Coordenades'].str.replace(' ', ',')
vitrallOutput
```

```python
vitrallOutput[['Coordenades']].to_csv(path_or_buf="out.csv", encoding='latin1')
```
